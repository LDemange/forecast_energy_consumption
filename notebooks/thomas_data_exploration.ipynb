```python
import pandas as pd 
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df_weather = pd.read_csv("../raw_data/weather_power_nasa.csv")
df_weather['date'] = df_weather['date'].apply(lambda _: datetime.strptime(_,"%Y-%m-%d %H:%M:%S"))
df_weather = df_weather.rename(columns = {'date': 'date_Y_M_D_H'})
df_weather.head()
#df_weather['date_Y_M_D_H']=df_weather['date']
```

```python
df_train = pd.read_csv("../raw_data/train.csv")
#df_train.head()
```

```python
df_train['date']=df_train['date'].apply(lambda _: datetime.strptime(_,"%Y-%m-%d %H:%M:%S"))
```

```python
df_train.dtypes
```

```python
df_train['year'] = df_train['date'].dt.strftime('%Y')
df_train['month'] = df_train['date'].dt.strftime('%m')
df_train['day'] = df_train['date'].dt.strftime('%d')
df_train['hour'] = df_train['date'].dt.strftime('%H')
df_train['minute'] = df_train['date'].dt.strftime('%M')
df_train['date_Y_M_D'] = df_train['date'].dt.strftime('%Y-%m-%d').apply(lambda _: datetime.strptime(_,"%Y-%m-%d"))
df_train['date_Y_M_D_H'] = df_train['date'].dt.strftime('%Y-%m-%d %H').apply(lambda _: datetime.strptime(_,"%Y-%m-%d %H"))
df_train.dtypes
```

```python
df_train
```

```python
df_merged = pd.merge(df_train, df_weather, on=["date_Y_M_D_H","metropolitan_area_code"], how="inner").reset_index(drop=True)

df_merged
```

```python
df_merged.to_csv('../raw_data/df_merged.csv')
```

```python
import pandas as pd

l = (pd.DataFrame(columns=['NULL'],
                  index=pd.date_range('2017-02-13 02:30:00', '2021-12-31 23:45:00',
                                      freq='15T'))
       .index.strftime('%Y-%m-%d %H:%M:%S')
       .tolist()
)
```

```python
l
```

```python
df_area_date = pd.DataFrame( columns = ['metropolitan_area_code', 'date'])
```

```python
df_new = pd.DataFrame( columns = ['metropolitan_area_code', 'date'])
```

```python
for i in range(0,17):
    df_new["metropolitan_area_code"] = i
    
    df_new["date"] = l

    df_area_date = pd.concat([df_area_date,df_new], ignore_index=True)
    
    
```

```python
df_area_date
```

```python
df_area_date['metropolitan_area_code'] = df_area_date['metropolitan_area_code'].fillna(0)
df_area_date['date'] = df_area_date['date'].apply(lambda _: datetime.strptime(_,"%Y-%m-%d %H:%M:%S"))
```

```python
df_area_date.to_csv('../raw_data/df_area_date.csv')
```
